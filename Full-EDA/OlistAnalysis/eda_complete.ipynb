## Recommendation using FP-Growth

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules

df_sales = df.groupby("order_id")["product_category_name_english"].apply(list).to_list()
df_sales
```

```python
te = TransactionEncoder()
te_done = te.fit(df_sales).transform(df_sales)
done_df = pd.DataFrame(te_done, columns = te.columns_)
growth = fpgrowth(done_df, min_support = 0.0001, use_colnames = True)
growth
```

```python
pairs = association_rules(growth, min_threshold = 0.05, metric = "lift")
pairs = pairs.sort_values("lift", ascending = False)
pairs
```

```python
pairs["antecedents"] = pairs["antecedents"].apply(lambda x: ", ".join(list(x))).astype("unicode")
pairs["consequents"] = pairs["consequents"].apply(lambda x: ", ".join(list(x))).astype("unicode")
pairs
```

### Recommendation Results

```python
import networkx as nx

pairs_top = pairs.head(300)

G = nx.from_pandas_edgelist(
    pairs_top,
    source = "antecedents",
    target = "consequents",
    edge_attr = "lift"
)

plt.figure(figsize = (10, 5))
pos = nx.spring_layout(G, k = 0.2, seed = 42)

nx.draw(
    G,
    pos,
    with_labels = True,
    node_size = 500,
    width = [G[u][v]["lift"] for u, v in G.edges()]
)
plt.title("Graph Plot of Product Clusters")
```

## Reviews NLP

```python
df_nlp = df[["order_id", "product_category_name_english", "review_score", "review_comment_message"]].dropna(subset = ["review_comment_message"]).reset_index()
df_nlp = df_nlp.drop("index", axis = 1)
df_nlp
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, is_torch_available
import torch.nn as nn
print(is_torch_available())
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-multilingual-uncased")
model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-multilingual-uncased")
```

```python
df_nlp["score"] = df_nlp["review_score"] >= 3 
df_nlp["score"] = df_nlp["score"].astype(int)

df_nlp
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_nlp, test_size = .3, random_state = 42)
df_train = df_train.reset_index()
df_train = df_train.drop("index", axis = 1)
df_test = df_test.reset_index()
df_test = df_test.drop("index", axis = 1)
df_train.head()
```

### Tokenizing inputs

```python
inputs_train = tokenizer(
    df_train["review_comment_message"].tolist(),
    do_lower_case = True,
    padding = True,
    trucation = True,
    max_lenght = 128,
    return_tensors = "pt"
)
inputs_test = tokenizer(
    df_test["review_comment_message"].tolist(),
    do_lower_case = True,
    padding = True,
    trucation = True,
    max_lenght = 128,
    return_tensors = "pt"
)

inputs_train
```

### Creating DataLoaders

```python
import torch
from torch.utils.data import Dataset, DataLoader

class dataset(Dataset):

    def __init__(self, dataset, label):
        self.dataset = dataset
        self.label = label

    def __len__(self):
        return len(self.dataset["input_ids"])
    
    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        sample = {
            "input_ids" : self.dataset["input_ids"][idx],
            "attention_mask" : self.dataset["attention_mask"][idx],
            "labels" : torch.tensor(self.label[idx])
        }
        return sample

data_train = dataset(inputs_train, df_train["score"])
data_test = dataset(inputs_test, df_test["score"])

dataload_train = DataLoader(
    data_train,
    batch_size = 32,
    shuffle = True,
    num_workers = 0
)
dataload_test = DataLoader(
    data_test,
    batch_size = 32,
    shuffle = True,
    num_workers = 0
)
print(len(data_train))
print(len(data_test))
```

### Defining Class Weights for unbalanced data

```python
from torch.optim import AdamW
from torch.nn import CrossEntropyLoss
from torch import Tensor
from torch.cuda import device

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

classes = df_train["score"].value_counts(normalize = True)

weight_zero = 1 / classes[0]
weight_one = 1 / classes[1]

ce = CrossEntropyLoss(weight = Tensor([weight_zero, weight_one]).to(device))
optimizers_params = [
    {"params" : model.bert.parameters(), "lr" : 0.00001},
    {"params" : model.classifier.parameters(), "lr" : 0.0001}
]

# optimizer = AdamW(model.parameters(), lr = 0.0001)
optimizer = AdamW(optimizers_params)
```

```python
for name, param in model.named_parameters():
    print(f"{name} is trainable: {param.requires_grad}")
```

### Complete Fine-Tuning of Bert Classifier

```python
from tqdm import tqdm
from datetime import datetime
from torch.utils.tensorboard.writer import SummaryWriter

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
writer = SummaryWriter("runs/fashion_trainer_{}".format(timestamp))

model.to(device)

epochs = 5
best_val_loss = 10000.0


for epoch in range(1, epochs + 1):
    print(f"Starting Epoch {epoch} / {epochs}")
    pbar = tqdm(total = int(len(dataload_train)), unit_scale = 1, desc = "...")

    model.train(True)
    loss_value = 0.0
    total_train_loss = 0.0

    for i, data in enumerate(dataload_train):
        inputs = {key: value.to(device) for key, value in data.items()}

        optimizer.zero_grad()

        output = model(**inputs)

        loss = ce(output.logits, inputs["labels"])
        loss.backward()

        optimizer.step()

        loss_value = loss.item()
        total_train_loss += loss_value
        avg_train_loss = total_train_loss / (i + 1)
        
        tb_x = epoch * len(dataload_train) + i + 1
        writer.add_scalar("Loss/train_avg", avg_train_loss, tb_x)

        tqdm_desc = f"Train Loss: {avg_train_loss:.4f}"
        pbar.set_description(desc = tqdm_desc)

        loss_value = 0.0

        pbar.update(1)

    pbar.close()
    
    current_val_loss = 0.0
    model.eval()
    print("Evaluation...")
    ebar = tqdm(total = int(len(dataload_test)), unit_scale = 1, desc = "...")
    with torch.no_grad():
        for i, vdata in enumerate(dataload_test):
            val_inputs = {key: value.to(device) for key, value in vdata.items()}
            val_outputs = model(**val_inputs)
            # vloss = val_outputs.loss
            vloss = ce(val_outputs.logits, val_inputs["labels"])
            current_val_loss += vloss.item()

            avg_val_loss = current_val_loss / (i + 1)
            tqdm_desc = f"Test Loss: {avg_val_loss:.4f}"
            ebar.set_description(desc = tqdm_desc)
            ebar.update(1)
        ebar.close()

    writer.add_scalars('Training vs. Validation Loss',
                    { 'Training' : avg_train_loss, 'Validation' : avg_val_loss},
                    epoch + 1)
    writer.flush()
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        model_path = "model_results/model_{}_{}".format(timestamp, epoch)
        torch.save(model.state_dict(), model_path)

print("Training finished.")
```

```python
from tqdm import tqdm

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    tbar = tqdm(total = int(len(dataload_test)), unit_scale = 1, desc = "...")
    for i, data in enumerate(dataload_test):
        inputs = {key: value.to(device) for key, value in data.items() if key != "labels"}

        labels = data["labels"].to(device)

        out = model(**inputs)
        
        preds = torch.argmax(out.logits, dim = -1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        tbar.update(1)

y_true = np.array(all_labels)
y_pred = np.array(all_preds)
```

### SHAP Analysis for the Bert Classifier

```python
import shap
print("Starting")
def wrap(x):
    tv = tokenizer(
        x.tolist(),
        do_lower_case = True,
        padding = True,
        trucation = True,
        max_lenght = 128,
        return_tensors = "pt"
    )

    inputs = {key: value.to(device) for key, value in tv.items()}
    with torch.no_grad():
        outputs = model(**inputs)
        scores = torch.nn.functional.softmax(outputs.logits, dim = -1)
        return scores.cpu().numpy()


masker = shap.maskers.Text(tokenizer, mask_token = "[MASK]")
explainer = shap.Explainer(wrap, masker, output_names = ["Neg", "Pos"])
shap_values = explainer(df_test["review_comment_message"])
```

```python
import pickle
with open("shap_values_olist.pkl", "wb") as f:
    pickle.dump(shap_values, f)
```

### False Negatives SHAP Analysis

```python
import shap
error_indices = np.where((y_true == 1) & (y_pred == 0))[0] # False negatives
shap_errors = shap_values[error_indices]

shap.plots.bar(shap_errors[:, :, 1].mean(0))
```

False negatives are often driven by strongly negative lexical cues (e.g., "nao", "diferente"), which dominate the model’s decision even when the overall rating is positive.

### False Positives SHAP Analysis

```python
fp_indices = np.where((y_true == 0) & (y_pred == 1))[0] # False positives
shap_fp = shap_values[fp_indices]

shap.plots.bar(shap_fp[:, :, 1].mean(0))
```

This shows that misclassifications are mostly driven by strong polarity tokens inside mixed-sentiment reviews, rather than random noise. Making the model rarely mismatch the correct class, but doesn't make it impossible.
