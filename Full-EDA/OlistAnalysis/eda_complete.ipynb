```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
df = pd.read_csv("completed_datasets/eda_olist_complete.csv", sep = ",")
df.head()
```

```python
list_of_times = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_customer_date",
    "review_creation_date",
    "review_answer_timestamp"
]
for item in list_of_times:
    df[item] = pd.to_datetime(df[item])
    
df.head()
```

```python
df_top_price = df.groupby("customer_state").sum(numeric_only = True).sort_values(by = "payment_value", ascending = False).head(10).reset_index()

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize = (4, 4))

sns.barplot(
    data = df_top_price,
    x = "payment_value",
    y = "customer_state",
    palette = "magma",
    hue = "payment_value",
    legend = False
)
plt.ylabel("") 
plt.xlabel("Payment Value")
plt.title("Top 10 States with most payment value")
```

```python
df_top_sellers = df.groupby("product_category_name_english").sum(numeric_only = True).sort_values(by = "payment_value", ascending = False).head(10).reset_index()
```

```python
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize = (4, 4))

sns.barplot(
    data = df_top_sellers,
    x = "payment_value",
    y = "product_category_name_english",
    palette = "magma",
    hue = "payment_value",
    legend = False
)
plt.ylabel("") 
plt.xlabel("Payment Value")
plt.title("Top 10 Product Categories by payment value")
```

```python
df_best_category = df.groupby("product_category_name_english").sum(numeric_only = True).sort_values(by = "review_score", ascending = False).head(10).reset_index()
df_best_category.head()

sns.set_theme(style = "whitegrid")
fig, ax = plt.subplots(figsize = (4, 4))

sns.barplot(
    data = df_best_category,
    x = "review_score",
    y = "product_category_name_english",
    palette = "magma",
    hue = "payment_value",
    legend = False
)
plt.ylabel("") 
plt.xlabel("Total reviews scores")
plt.title("Top 10 Product Categories by reviews")
```

```python
df_most_popular_state = df["seller_state"].value_counts().head(10).reset_index()
sns.set_theme(style = "whitegrid")
fig, ax = plt.subplots(figsize = (4, 4))

sns.barplot(
    data = df_most_popular_state,
    x = "count",
    y = "seller_state",
    palette = "magma",
    hue = "count",
    legend = False
)
plt.title("Top 10 Popular States")
```

```python
fidelity_df = df[["order_purchase_timestamp", "customer_unique_id"]]
fidelity_df["order_purchase_timestamp"] = fidelity_df["order_purchase_timestamp"].dt.date

fidelity_df = fidelity_df.drop_duplicates(subset = ["customer_unique_id", "order_purchase_timestamp"])

returns = (fidelity_df["customer_unique_id"].value_counts() > 1).sum()

sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(2, 1, figsize = (4, 4))

labels = ["Non-returning", "Returning"]
counts = [len(fidelity_df) - returns, returns]

ax[0].pie(
    x = counts,
    labels = labels,
    autopct = "%1.1f%%"
)
ax[0].set_title("Pie chart of returning customers")


counts = fidelity_df["customer_unique_id"].value_counts()

counts[counts > 1].value_counts().plot(kind = "bar", color = "purple")
plt.xlabel("Total Orders")
plt.ylabel("Number of Clients")
plt.title("How many times a client returned?")
```

```python
id_fidelity = counts[counts > 1].index
df_fidelity = df[df["customer_unique_id"].isin(id_fidelity)]

loyal_amount = df_fidelity["payment_value"].sum()
total_amount = df["payment_value"].sum()

print(f"Total value from loyal clients: {loyal_amount:.2f}")
print(f"Total value from total clients: {total_amount:.2f}")
print(f"Loyal clients make up for the {(loyal_amount / total_amount) * 100:.2f}% of Payment Value")
```

```python
df_popular_payment_type = df["payment_type"].value_counts().head(10).reset_index()
df_popular_payment_installments = df["payment_installments"].value_counts().head(10).reset_index()

sns.set_theme(style = "whitegrid")
fig, ax = plt.subplots(2, 1, figsize = (6, 6))

ax[0].pie(
    labels = df_popular_payment_type["payment_type"],
    x = df_popular_payment_type["count"],
    autopct = "%1.1f%%"
)
ax[0].set_title("Popularity of Payment Type")

sns.barplot(
    data = df_popular_payment_installments,
    x = "payment_installments",
    y = "count",
    palette = "magma",
    hue = "count",
    legend = False,
    ax = ax[1]
)
ax[1].set_title("Popularity of Payment Installments")
ax[1].set_ylabel("Count")
ax[1].set_xlabel("Installments")
```
